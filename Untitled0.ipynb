```python
import json
import spacy
from spacy.tokens import DocBin

nlp = spacy.blank("en")

def is_valid_span(doc, start, end):
    span = doc.char_span(start, end)
    return span is not None

def jsonl_to_spacy(jsonl_path, output_path):
    db = DocBin()
    skipped = 0

    with open(jsonl_path, "r", encoding="utf-8") as f:
        for line in f:
            item = json.loads(line)
            text = item["text"]
            labels = item.get("label", {})

            doc = nlp.make_doc(text)
            ents = []

            for key, value in labels.items():

                # LIST LABELS (Skills list)
                if isinstance(value, list):
                    for val in value:
                        idx = text.lower().find(val.lower())
                        if idx != -1:
                            start = idx
                            end = idx + len(val)
                            if is_valid_span(doc, start, end):
                                ents.append(doc.char_span(start, end, label=key))
                            else:
                                skipped += 1

                # STRING LABELS
                elif isinstance(value, str):
                    idx = text.lower().find(value.lower())
                    if idx != -1:
                        start = idx
                        end = idx + len(value)
                        if is_valid_span(doc, start, end):
                            span = doc.char_span(start, end, label=key)
                            ents.append(span)
                        else:
                            skipped += 1

            # Remove overlaps
            filtered = []
            occupied = set()
            for ent in sorted(ents, key=lambda x: (x.start_char, x.end_char)):
                if ent is None:
                    continue
                rng = set(range(ent.start_char, ent.end_char))
                if occupied.intersection(rng):
                    skipped += 1
                    continue
                occupied.update(rng)
                filtered.append(ent)

            doc.ents = filtered
            db.add(doc)

    db.to_disk(output_path)
    print("Saved:", output_path)
    print("Skipped spans:", skipped)
```

```python
jsonl_to_spacy("data.jsonl", "train.spacy")
```

```python
import spacy
from spacy.scorer import Scorer
from spacy.training.example import Example
from spacy.tokens import DocBin

# Load trained model
nlp = spacy.load("/content/output/model-best")

# Load DEV data
dev_data = DocBin().from_disk("train.spacy")
docs = list(dev_data.get_docs(nlp.vocab))

scorer = Scorer()

examples = []
for gold_doc in docs:
    pred_doc = nlp(gold_doc.text)
    example = Example(pred_doc, gold_doc)
    examples.append(example)

scores = scorer.score(examples)
print("REAL ACCURACY METRICS:\n")
print("Precision:", scores["ents_p"])
print("Recall:", scores["ents_r"])
print("F1 Score:", scores["ents_f"])
print("\nPer-Label Scores:")
print(scores["ents_per_type"])
```
